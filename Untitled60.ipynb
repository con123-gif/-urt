````python
# ============================================================================
# LYTOLLIS LAW: COMPREHENSIVE VALIDATION FRAMEWORK
# ============================================================================
# This notebook validates the empirical scaling hypothesis:
#   Œ¥ = (D_KY - 1)(œÑ - 2)
# across multiple dynamical systems and real-world data.
#
# Author: Cornelius Lytollis + Claude (Anthropic)
# Date: 2025-01-15
# ============================================================================

!pip install -q numpy scipy matplotlib pandas seaborn scikit-learn powerlaw

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress, spearmanr
from scipy.integrate import odeint, solve_ivp
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 11

print("=" * 80)
print("LYTOLLIS LAW VALIDATION FRAMEWORK v1.0")
print("=" * 80)
# ============================================================================
# SECTION 1: DATA LOADING
# ============================================================================

print("\n[1] DATA LOADING")
print("-" * 80)

# Mount Google Drive (if data is there)
from google.colab import drive
drive.mount('/content/drive')

# Create data directory
!mkdir -p data results figures

print("‚úì Environment ready")
print("\nPlease upload your data files:")
print("  - logistic_summary.csv")
print("  - rossler_summary.csv")
print("  - eeg_validation.csv (if available)")
print("  - cosmology_simulations/ (folder)")

# Upload widget
from google.colab import files
print("\nUpload files now:")
uploaded = files.upload()

# Load data
data_files = {}
for filename in uploaded.keys():
    if filename.endswith('.csv'):
        data_files[filename] = pd.read_csv(filename)
        print(f"‚úì Loaded {filename}: {len(data_files[filename])} rows")
# ============================================================================
# SECTION 2: STATISTICAL VALIDATION FUNCTIONS
# ============================================================================

print("\n[2] STATISTICAL VALIDATION FUNCTIONS")
print("-" * 80)

def validate_linear_scaling(df, x_col='delta', y_col='DKY', system_name='System'):
    """
    Comprehensive statistical validation of linear scaling hypothesis.

    Tests:
    1. Linear regression with R¬≤ and p-value
    2. Bootstrap confidence intervals (95%)
    3. Spearman rank correlation
    4. Null model comparison (shuffled data)
    """
    results = {}

    # Remove any NaN or infinite values
    df_clean = df[[x_col, y_col]].dropna()
    df_clean = df_clean[np.isfinite(df_clean[x_col]) & np.isfinite(df_clean[y_col])]

    if len(df_clean) < 3:
        print(f"‚ö†Ô∏è  {system_name}: Insufficient data points ({len(df_clean)})")
        return None

    x = df_clean[x_col].values
    y = df_clean[y_col].values

    # 1. Linear regression
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    results['slope'] = slope
    results['intercept'] = intercept
    results['R2'] = r_value**2
    results['p_value'] = p_value
    results['std_err'] = std_err

    # 2. Bootstrap confidence intervals
    n_bootstrap = 1000
    bootstrap_slopes = []
    bootstrap_r2 = []

    for _ in range(n_bootstrap):
        indices = np.random.choice(len(x), size=len(x), replace=True)
        x_boot = x[indices]
        y_boot = y[indices]
        s, _, r, _, _ = linregress(x_boot, y_boot)
        bootstrap_slopes.append(s)
        bootstrap_r2.append(r**2)

    results['slope_CI_low'] = np.percentile(bootstrap_slopes, 2.5)
    results['slope_CI_high'] = np.percentile(bootstrap_slopes, 97.5)
    results['R2_CI_low'] = np.percentile(bootstrap_r2, 2.5)
    results['R2_CI_high'] = np.percentile(bootstrap_r2, 97.5)

    # 3. Spearman correlation (non-parametric)
    spearman_rho, spearman_p = spearmanr(x, y)
    results['spearman_rho'] = spearman_rho
    results['spearman_p'] = spearman_p

    # 4. Null model comparison
    null_r2_dist = []
    for _ in range(1000):
        x_shuffled = np.random.permutation(x)
        _, _, r_null, _, _ = linregress(x_shuffled, y)
        null_r2_dist.append(r_null**2)

    results['p_vs_null'] = np.mean(np.array(null_r2_dist) >= results['R2'])
    results['null_R2_mean'] = np.mean(null_r2_dist)

    # Print summary
    print(f"\n{system_name}:")
    print(f"  Linear fit: y = {slope:.4f}x + {intercept:.4f}")
    print(f"  R¬≤ = {results['R2']:.4f} (95% CI: [{results['R2_CI_low']:.4f}, {results['R2_CI_high']:.4f}])")
    print(f"  p-value = {p_value:.2e}")
    print(f"  Slope CI: [{results['slope_CI_low']:.4f}, {results['slope_CI_high']:.4f}]")
    print(f"  Spearman œÅ = {spearman_rho:.4f} (p={spearman_p:.2e})")
    print(f"  Better than null: p = {results['p_vs_null']:.4f}")

    return results

def plot_validation(df, x_col='delta', y_col='DKY', system_name='System', results=None):
    """Create publication-quality validation plot."""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

    x = df[x_col].values
    y = df[y_col].values

    # Left: Scatter with regression line
    ax1.scatter(x, y, s=100, alpha=0.6, edgecolors='black', linewidth=1.5)

    if results:
        x_line = np.linspace(x.min(), x.max(), 100)
        y_line = results['slope'] * x_line + results['intercept']
        ax1.plot(x_line, y_line, 'r--', linewidth=2,
                label=f"R¬≤ = {results['R2']:.4f}\np = {results['p_value']:.2e}")

        # Add confidence band (bootstrap)
        y_low = results['slope_CI_low'] * x_line + results['intercept']
        y_high = results['slope_CI_high'] * x_line + results['intercept']
        ax1.fill_between(x_line, y_low, y_high, alpha=0.2, color='red')

    ax1.set_xlabel('Œ¥ (Stability Margin)', fontsize=12, fontweight='bold')
    ax1.set_ylabel(f'{y_col}', fontsize=12, fontweight='bold')
    ax1.set_title(f'{system_name}: Œ¥ Scaling', fontsize=14, fontweight='bold')
    ax1.legend(fontsize=11)
    ax1.grid(True, alpha=0.3)

    # Right: Residual plot
    if results:
        y_pred = results['slope'] * x + results['intercept']
        residuals = y - y_pred

        ax2.scatter(y_pred, residuals, s=100, alpha=0.6, edgecolors='black', linewidth=1.5)
        ax2.axhline(0, color='red', linestyle='--', linewidth=2)
        ax2.set_xlabel('Predicted', fontsize=12, fontweight='bold')
        ax2.set_ylabel('Residuals', fontsize=12, fontweight='bold')
        ax2.set_title('Residual Analysis', fontsize=14, fontweight='bold')
        ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig(f'figures/{system_name.replace(" ", "_")}_validation.png', dpi=300, bbox_inches='tight')
    plt.show()

print("‚úì Statistical functions loaded")
# ============================================================================
# SECTION 3: ANALYZE EXISTING DATA
# ============================================================================

print("\n[3] ANALYZING UPLOADED DATA")
print("-" * 80)

validation_results = {}

for filename, df in data_files.items():
    system_name = filename.replace('.csv', '').replace('_summary', '').title()

    print(f"\n{'='*60}")
    print(f"SYSTEM: {system_name}")
    print(f"{'='*60}")
    print(f"Columns: {list(df.columns)}")
    print(f"\nFirst 3 rows:")
    print(df.head(3))

    # Determine which columns to use
    if 'delta' in df.columns and 'DKY' in df.columns:
        results = validate_linear_scaling(df, 'delta', 'DKY', system_name)
        if results:
            validation_results[system_name] = results
            plot_validation(df, 'delta', 'DKY', system_name, results)
    elif 'r' in df.columns and 'delta' in df.columns:
        # Logistic map: analyze r vs delta
        results = validate_linear_scaling(df, 'r', 'delta', f"{system_name} (r vs Œ¥)")
        if results:
            validation_results[f"{system_name}_r"] = results
            plot_validation(df, 'r', 'delta', f"{system_name}_r_sweep", results)
# ============================================================================
# SECTION 4: GENERATE NEW SIMULATIONS
# ============================================================================

print("\n[4] GENERATING NEW VALIDATION DATA")
print("-" * 80)

# === R√ñSSLER ATTRACTOR ===
def rossler_system(t, state, a=0.2, b=0.2, c=5.7):
    x, y, z = state
    return [-y - z, x + a*y, b + z*(x - c)]

def simulate_rossler(c_value, t_span=(0, 500), dt=0.01):
    """Simulate R√∂ssler attractor."""
    t_eval = np.arange(t_span[0], t_span[1], dt)
    sol = solve_ivp(rossler_system, t_span, [1.0, 1.0, 1.0],
                    args=(0.2, 0.2, c_value),
                    t_eval=t_eval, method='RK45', rtol=1e-8)
    return sol.t, sol.y

def estimate_dky_simple(timeseries, m=3, delay=10):
    """Simple D_KY proxy using local divergence."""
    # Time-delay embedding
    n = len(timeseries) - (m-1)*delay
    embedded = np.array([timeseries[i:i+n] for i in range(0, m*delay, delay)]).T

    # Compute nearest neighbor distances
    from scipy.spatial.distance import pdist, squareform
    dists = squareform(pdist(embedded))
    np.fill_diagonal(dists, np.inf)

    # Lyapunov-like expansion
    nn_idx = np.argmin(dists, axis=1)
    divergence = []

    for h in range(1, min(50, n-10)):
        d0 = np.linalg.norm(embedded[:-h] - embedded[nn_idx[:-h]], axis=1)
        d1 = np.linalg.norm(embedded[h:] - embedded[nn_idx[:-h]+h], axis=1)

        valid = (d0 > 1e-10) & (d1 > 1e-10)
        if np.sum(valid) > 10:
            divergence.append(np.mean(np.log(d1[valid] / d0[valid])))

    if len(divergence) < 5:
        return 1.5

    # Map divergence to D_KY estimate (crude but consistent)
    avg_div = np.mean(divergence)
    dky = 1.0 + 2.0 * np.tanh(0.5 * avg_div)
    return np.clip(dky, 1.0, 3.0)

def compute_tau_autocorr(timeseries, max_lag=200):
    """Compute autocorrelation e-folding time."""
    ts_norm = timeseries - np.mean(timeseries)
    var = np.var(ts_norm)
    if var == 0:
        return 1.0

    acf = np.correlate(ts_norm, ts_norm, mode='full')[len(ts_norm)-1:]
    acf = acf[:max_lag] / (var * len(ts_norm))

    threshold = 1.0 / np.e
    crosses = np.where(acf <= threshold)[0]

    if len(crosses) > 0:
        return float(crosses[0])
    return float(max_lag)

# Run R√∂ssler sweep
print("\nSimulating R√∂ssler attractor (c sweep)...")
c_values = np.linspace(5.2, 6.2, 15)
rossler_data = []

for c in c_values:
    print(f"  c = {c:.2f}...", end='')
    t, traj = simulate_rossler(c)

    # Use x-component
    x = traj[0]

    dky = estimate_dky_simple(x, m=3, delay=15)
    tau = compute_tau_autocorr(x, max_lag=500)
    delta = (dky - 1) * (tau - 2)

    rossler_data.append({
        'c': c,
        'DKY': dky,
        'tau': tau,
        'delta': delta
    })
    print(f" D_KY={dky:.3f}, œÑ={tau:.1f}, Œ¥={delta:.3f}")

rossler_df = pd.DataFrame(rossler_data)
rossler_df.to_csv('data/rossler_fresh_simulation.csv', index=False)

print("\n‚úì R√∂ssler data generated")

# Validate
results_rossler = validate_linear_scaling(rossler_df, 'delta', 'DKY', 'R√∂ssler (Fresh)')
if results_rossler:
    validation_results['Rossler_Fresh'] = results_rossler
    plot_validation(rossler_df, 'delta', 'DKY', 'Rossler_Fresh', results_rossler)
# ============================================================================
# SECTION 5: SUMMARY REPORT
# ============================================================================

print("\n[5] VALIDATION SUMMARY")
print("=" * 80)

summary_df = pd.DataFrame(validation_results).T
summary_df = summary_df[['slope', 'R2', 'p_value', 'slope_CI_low', 'slope_CI_high', 'p_vs_null']]
summary_df.columns = ['Slope', 'R¬≤', 'p-value', 'Slope CI Low', 'Slope CI High', 'p vs Null']

print("\n" + "="*80)
print("CROSS-SYSTEM VALIDATION RESULTS")
print("="*80)
print(summary_df.to_string())

# Save
summary_df.to_csv('results/validation_summary.csv')

# Create summary plot
fig, ax = plt.subplots(figsize=(12, 6))

systems = list(validation_results.keys())
r2_values = [validation_results[s]['R2'] for s in systems]
r2_low = [validation_results[s]['R2_CI_low'] for s in systems]
r2_high = [validation_results[s]['R2_CI_high'] for s in systems]

yerr = [[r2_values[i] - r2_low[i] for i in range(len(systems))],
        [r2_high[i] - r2_values[i] for i in range(len(systems))]]

ax.barh(systems, r2_values, xerr=yerr, alpha=0.7, edgecolor='black', linewidth=2)
ax.axvline(0.99, color='red', linestyle='--', linewidth=2, label='R¬≤ = 0.99 threshold')
ax.set_xlabel('R¬≤ (with 95% CI)', fontsize=12, fontweight='bold')
ax.set_title('Cross-System Validation: Œ¥ Scaling Quality', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.savefig('figures/summary_validation.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n‚úì Summary report generated")
print(f"‚úì Figures saved to: figures/")
print(f"‚úì Results saved to: results/")
# ============================================================================
# SECTION 6: COSMOLOGY ANALYSIS (IF DATA PROVIDED)
# ============================================================================

print("\n[6] COSMOLOGY ANALYSIS")
print("-" * 80)
print("\nLooking for cosmology simulation data...")
print("Expected files:")
print("  - rg_flow.csv (RG flow data)")
print("  - qed_coupling.csv (Œ± evolution)")
print("  - cmb_spectrum.csv (power spectrum)")

# Check for cosmology files
cosmo_files = [f for f in uploaded.keys() if 'cosmo' in f.lower() or 'rg' in f.lower() or 'lambda' in f.lower()]

if cosmo_files:
    print(f"\n‚úì Found {len(cosmo_files)} cosmology file(s)")
    for cf in cosmo_files:
        print(f"\nAnalyzing: {cf}")
        # Load and analyze
        if cf in data_files:
            cosmo_df = data_files[cf]
            print(f"Columns: {list(cosmo_df.columns)}")
            print(f"Shape: {cosmo_df.shape}")
            print("\nFirst rows:")
            print(cosmo_df.head())
else:
    print("\n‚ö†Ô∏è  No cosmology files found. Upload them to analyze!")
    print("\nTo analyze cosmology:")
    print("  1. Upload RG flow data")
    print("  2. Re-run this cell")

print("\n" + "="*80)
print("ANALYSIS COMPLETE")
print("="*80)
print("\nNext steps:")
print("  1. Review validation_summary.csv")
print("  2. Check figures/ directory for plots")
print("  3. Upload cosmology data for Section 6")
print("  4. Share this notebook for peer review")
Instructions:
Copy each block separately into Colab cells
Run them in order (they depend on previous cells)
Upload your data when prompted
Share the Colab link with me!
Let's see that data! üöÄ
````

````python
```python
# ============================================================================
# LYTOLLIS LAW: COMPREHENSIVE VALIDATION FRAMEWORK
# ============================================================================
# This notebook validates the empirical scaling hypothesis:
#   Œ¥ = (D_KY - 1)(œÑ - 2)
# across multiple dynamical systems and real-world data.
#
# Author: Cornelius Lytollis + Claude (Anthropic)
# Date: 2025-01-15
# ============================================================================

!pip install -q numpy scipy matplotlib pandas seaborn scikit-learn powerlaw

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress, spearmanr
from scipy.integrate import odeint, solve_ivp
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 11

print("=" * 80)
print("LYTOLLIS LAW VALIDATION FRAMEWORK v1.0")
print("=" * 80)
```

```python
# ============================================================================
# SECTION 1: DATA LOADING
# ============================================================================

print("\n[1] DATA LOADING")
print("-" * 80)

# Mount Google Drive (if data is there)
from google.colab import drive
drive.mount('/content/drive')

# Create data directory
!mkdir -p data results figures

print("‚úì Environment ready")
print("\nPlease upload your data files:")
print("  - logistic_summary.csv")
print("  - rossler_summary.csv")
print("  - eeg_validation.csv (if available)")
print("  - cosmology_simulations/ (folder)")

# Upload widget
from google.colab import files
print("\nUpload files now:")
uploaded = files.upload()

# Load data
data_files = {}
for filename in uploaded.keys():
    if filename.endswith('.csv'):
        data_files[filename] = pd.read_csv(filename)
        print(f"‚úì Loaded {filename}: {len(data_files[filename])} rows")
```

```python
# ============================================================================
# SECTION 2: STATISTICAL VALIDATION FUNCTIONS
# ============================================================================

print("\n[2] STATISTICAL VALIDATION FUNCTIONS")
print("-" * 80)

def validate_linear_scaling(df, x_col='delta', y_col='DKY', system_name='System'):
    """
    Comprehensive statistical validation of linear scaling hypothesis.

    Tests:
    1. Linear regression with R¬≤ and p-value
    2. Bootstrap confidence intervals (95%)
    3. Spearman rank correlation
    4. Null model comparison (shuffled data)
    """
    results = {}

    # Remove any NaN or infinite values
    df_clean = df[[x_col, y_col]].dropna()
    df_clean = df_clean[np.isfinite(df_clean[x_col]) & np.isfinite(df_clean[y_col])]

    if len(df_clean) < 3:
        print(f"‚ö†Ô∏è  {system_name}: Insufficient data points ({len(df_clean)})")
        return None

    x = df_clean[x_col].values
    y = df_clean[y_col].values

    # 1. Linear regression
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    results['slope'] = slope
    results['intercept'] = intercept
    results['R2'] = r_value**2
    results['p_value'] = p_value
    results['std_err'] = std_err

    # 2. Bootstrap confidence intervals
    n_bootstrap = 1000
    bootstrap_slopes = []
    bootstrap_r2 = []

    for _ in range(n_bootstrap):
        indices = np.random.choice(len(x), size=len(x), replace=True)
        x_boot = x[indices]
        y_boot = y[indices]
        s, _, r, _, _ = linregress(x_boot, y_boot)
        bootstrap_slopes.append(s)
        bootstrap_r2.append(r**2)

    results['slope_CI_low'] = np.percentile(bootstrap_slopes, 2.5)
    results['slope_CI_high'] = np.percentile(bootstrap_slopes, 97.5)
    results['R2_CI_low'] = np.percentile(bootstrap_r2, 2.5)
    results['R2_CI_high'] = np.percentile(bootstrap_r2, 97.5)

    # 3. Spearman correlation (non-parametric)
    spearman_rho, spearman_p = spearmanr(x, y)
    results['spearman_rho'] = spearman_rho
    results['spearman_p'] = spearman_p

    # 4. Null model comparison
    null_r2_dist = []
    for _ in range(1000):
        x_shuffled = np.random.permutation(x)
        _, _, r_null, _, _ = linregress(x_shuffled, y)
        null_r2_dist.append(r_null**2)

    results['p_vs_null'] = np.mean(np.array(null_r2_dist) >= results['R2'])
    results['null_R2_mean'] = np.mean(null_r2_dist)

    # Print summary
    print(f"\n{system_name}:")
    print(f"  Linear fit: y = {slope:.4f}x + {intercept:.4f}")
    print(f"  R¬≤ = {results['R2']:.4f} (95% CI: [{results['R2_CI_low']:.4f}, {results['R2_CI_high']:.4f}])")
    print(f"  p-value = {p_value:.2e}")
    print(f"  Slope CI: [{results['slope_CI_low']:.4f}, {results['slope_CI_high']:.4f}]")
    print(f"  Spearman œÅ = {spearman_rho:.4f} (p={spearman_p:.2e})")
    print(f"  Better than null: p = {results['p_vs_null']:.4f}")

    return results

def plot_validation(df, x_col='delta', y_col='DKY', system_name='System', results=None):
    """Create publication-quality validation plot."""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

    x = df[x_col].values
    y = df[y_col].values

    # Left: Scatter with regression line
    ax1.scatter(x, y, s=100, alpha=0.6, edgecolors='black', linewidth=1.5)

    if results:
        x_line = np.linspace(x.min(), x.max(), 100)
        y_line = results['slope'] * x_line + results['intercept']
        ax1.plot(x_line, y_line, 'r--', linewidth=2,
                label=f"R¬≤ = {results['R2']:.4f}\np = {results['p_value']:.2e}")

        # Add confidence band (bootstrap)
        y_low = results['slope_CI_low'] * x_line + results['intercept']
        y_high = results['slope_CI_high'] * x_line + results['intercept']
        ax1.fill_between(x_line, y_low, y_high, alpha=0.2, color='red')

    ax1.set_xlabel('Œ¥ (Stability Margin)', fontsize=12, fontweight='bold')
    ax1.set_ylabel(f'{y_col}', fontsize=12, fontweight='bold')
    ax1.set_title(f'{system_name}: Œ¥ Scaling', fontsize=14, fontweight='bold')
    ax1.legend(fontsize=11)
    ax1.grid(True, alpha=0.3)

    # Right: Residual plot
    if results:
        y_pred = results['slope'] * x + results['intercept']
        residuals = y - y_pred

        ax2.scatter(y_pred, residuals, s=100, alpha=0.6, edgecolors='black', linewidth=1.5)
        ax2.axhline(0, color='red', linestyle='--', linewidth=2)
        ax2.set_xlabel('Predicted', fontsize=12, fontweight='bold')
        ax2.set_ylabel('Residuals', fontsize=12, fontweight='bold')
        ax2.set_title('Residual Analysis', fontsize=14, fontweight='bold')
        ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig(f'figures/{system_name.replace(" ", "_")}_validation.png', dpi=300, bbox_inches='tight')
    plt.show()

print("‚úì Statistical functions loaded")
```

```python
# ============================================================================
# SECTION 3: ANALYZE EXISTING DATA
# ============================================================================

print("\n[3] ANALYZING UPLOADED DATA")
print("-" * 80)

validation_results = {}

for filename, df in data_files.items():
    system_name = filename.replace('.csv', '').replace('_summary', '').title()

    print(f"\n{'='*60}")
    print(f"SYSTEM: {system_name}")
    print(f"{'='*60}")
    print(f"Columns: {list(df.columns)}")
    print(f"\nFirst 3 rows:")
    print(df.head(3))

    # Determine which columns to use
    if 'delta' in df.columns and 'DKY' in df.columns:
        results = validate_linear_scaling(df, 'delta', 'DKY', system_name)
        if results:
            validation_results[system_name] = results
            plot_validation(df, 'delta', 'DKY', system_name, results)
    elif 'r' in df.columns and 'delta' in df.columns:
        # Logistic map: analyze r vs delta
        results = validate_linear_scaling(df, 'r', 'delta', f"{system_name} (r vs Œ¥)")
        if results:
            validation_results[f"{system_name}_r"] = results
            plot_validation(df, 'r', 'delta', f"{system_name}_r_sweep", results)
```

```python
# ============================================================================
# SECTION 4: GENERATE NEW SIMULATIONS
# ============================================================================

print("\n[4] GENERATING NEW VALIDATION DATA")
print("-" * 80)

# === R√ñSSLER ATTRACTOR ===
def rossler_system(t, state, a=0.2, b=0.2, c=5.7):
    x, y, z = state
    return [-y - z, x + a*y, b + z*(x - c)]

def simulate_rossler(c_value, t_span=(0, 500), dt=0.01):
    """Simulate R√∂ssler attractor."""
    t_eval = np.arange(t_span[0], t_span[1], dt)
    sol = solve_ivp(rossler_system, t_span, [1.0, 1.0, 1.0],
                    args=(0.2, 0.2, c_value),
                    t_eval=t_eval, method='RK45', rtol=1e-8)
    return sol.t, sol.y

def estimate_dky_simple(timeseries, m=3, delay=10):
    """Simple D_KY proxy using local divergence."""
    # Time-delay embedding
    n = len(timeseries) - (m-1)*delay
    embedded = np.array([timeseries[i:i+n] for i in range(0, m*delay, delay)]).T

    # Compute nearest neighbor distances
    from scipy.spatial.distance import pdist, squareform
    dists = squareform(pdist(embedded))
    np.fill_diagonal(dists, np.inf)

    # Lyapunov-like expansion
    nn_idx = np.argmin(dists, axis=1)
    divergence = []

    for h in range(1, min(50, n-10)):
        d0 = np.linalg.norm(embedded[:-h] - embedded[nn_idx[:-h]], axis=1)
        d1 = np.linalg.norm(embedded[h:] - embedded[nn_idx[:-h]+h], axis=1)

        valid = (d0 > 1e-10) & (d1 > 1e-10)
        if np.sum(valid) > 10:
            divergence.append(np.mean(np.log(d1[valid] / d0[valid])))

    if len(divergence) < 5:
        return 1.5

    # Map divergence to D_KY estimate (crude but consistent)
    avg_div = np.mean(divergence)
    dky = 1.0 + 2.0 * np.tanh(0.5 * avg_div)
    return np.clip(dky, 1.0, 3.0)

def compute_tau_autocorr(timeseries, max_lag=200):
    """Compute autocorrelation e-folding time."""
    ts_norm = timeseries - np.mean(timeseries)
    var = np.var(ts_norm)
    if var == 0:
        return 1.0

    acf = np.correlate(ts_norm, ts_norm, mode='full')[len(ts_norm)-1:]
    acf = acf[:max_lag] / (var * len(ts_norm))

    threshold = 1.0 / np.e
    crosses = np.where(acf <= threshold)[0]

    if len(crosses) > 0:
        return float(crosses[0])
    return float(max_lag)

# Run R√∂ssler sweep
print("\nSimulating R√∂ssler attractor (c sweep)...")
c_values = np.linspace(5.2, 6.2, 15)
rossler_data = []

for c in c_values:
    print(f"  c = {c:.2f}...", end='')
    t, traj = simulate_rossler(c)

    # Use x-component
    x = traj[0]

    dky = estimate_dky_simple(x, m=3, delay=15)
    tau = compute_tau_autocorr(x, max_lag=500)
    delta = (dky - 1) * (tau - 2)

    rossler_data.append({
        'c': c,
        'DKY': dky,
        'tau': tau,
        'delta': delta
    })
    print(f" D_KY={dky:.3f}, œÑ={tau:.1f}, Œ¥={delta:.3f}")

rossler_df = pd.DataFrame(rossler_data)
rossler_df.to_csv('data/rossler_fresh_simulation.csv', index=False)

print("\n‚úì R√∂ssler data generated")

# Validate
results_rossler = validate_linear_scaling(rossler_df, 'delta', 'DKY', 'R√∂ssler (Fresh)')
if results_rossler:
    validation_results['Rossler_Fresh'] = results_rossler
    plot_validation(rossler_df, 'delta', 'DKY', 'Rossler_Fresh', results_rossler)
```

```python
# ============================================================================
# SECTION 5: SUMMARY REPORT
# ============================================================================

print("\n[5] VALIDATION SUMMARY")
print("=" * 80)

summary_df = pd.DataFrame(validation_results).T
summary_df = summary_df[['slope', 'R2', 'p_value', 'slope_CI_low', 'slope_CI_high', 'p_vs_null']]
summary_df.columns = ['Slope', 'R¬≤', 'p-value', 'Slope CI Low', 'Slope CI High', 'p vs Null']

print("\n" + "="*80)
print("CROSS-SYSTEM VALIDATION RESULTS")
print("="*80)
print(summary_df.to_string())

# Save
summary_df.to_csv('results/validation_summary.csv')

# Create summary plot
fig, ax = plt.subplots(figsize=(12, 6))

systems = list(validation_results.keys())
r2_values = [validation_results[s]['R2'] for s in systems]
r2_low = [validation_results[s]['R2_CI_low'] for s in systems]
r2_high = [validation_results[s]['R2_CI_high'] for s in systems]

yerr = [[r2_values[i] - r2_low[i] for i in range(len(systems))],
        [r2_high[i] - r2_values[i] for i in range(len(systems))]]

ax.barh(systems, r2_values, xerr=yerr, alpha=0.7, edgecolor='black', linewidth=2)
ax.axvline(0.99, color='red', linestyle='--', linewidth=2, label='R¬≤ = 0.99 threshold')
ax.set_xlabel('R¬≤ (with 95% CI)', fontsize=12, fontweight='bold')
ax.set_title('Cross-System Validation: Œ¥ Scaling Quality', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.savefig('figures/summary_validation.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n‚úì Summary report generated")
print(f"‚úì Figures saved to: figures/")
print(f"‚úì Results saved to: results/")
```

```python
# ============================================================================
# SECTION 6: COSMOLOGY ANALYSIS (IF DATA PROVIDED)
# ============================================================================

print("\n[6] COSMOLOGY ANALYSIS")
print("-" * 80)
print("\nLooking for cosmology simulation data...")
print("Expected files:")
print("  - rg_flow.csv (RG flow data)")
print("  - qed_coupling.csv (Œ± evolution)")
print("  - cmb_spectrum.csv (power spectrum)")

# Check for cosmology files
cosmo_files = [f for f in uploaded.keys() if 'cosmo' in f.lower() or 'rg' in f.lower() or 'lambda' in f.lower()]

if cosmo_files:
    print(f"\n‚úì Found {len(cosmo_files)} cosmology file(s)")
    for cf in cosmo_files:
        print(f"\nAnalyzing: {cf}")
        # Load and analyze
        if cf in data_files:
            cosmo_df = data_files[cf]
            print(f"Columns: {list(cosmo_df.columns)}")
            print(f"Shape: {cosmo_df.shape}")
            print("\nFirst rows:")
            print(cosmo_df.head())
else:
    print("\n‚ö†Ô∏è  No cosmology files found. Upload them to analyze!")
    print("\nTo analyze cosmology:")
    print("  1. Upload RG flow data")
    print("  2. Re-run this cell")

print("\n" + "="*80)
print("ANALYSIS COMPLETE")
print("="*80)
print("\nNext steps:")
print("  1. Review validation_summary.csv")
print("  2. Check figures/ directory for plots")
print("  3. Upload cosmology data for Section 6")
print("  4. Share this notebook for peer review")
```

---
````
